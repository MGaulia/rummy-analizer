```python
from itertools import permutations 
```

```python
numbers = [
    'NUMBER_ACE',
    'NUMBER_TWO',
    'NUMBER_THREE',
    'NUMBER_FOUR',
    'NUMBER_FIVE',
    'NUMBER_SIX',
    'NUMBER_SEVEN',
    'NUMBER_EIGHT',
    'NUMBER_NINE',
    'NUMBER_TEN',
    'NUMBER_JACK',
    'NUMBER_QUEEN',
    'NUMBER_KING',
    'NUMBER_ACE'
]

colors = [
    'COLOR_BLACK',
    'COLOR_GREEN',
    'COLOR_GREY',
    'COLOR_PURPLE'
]
```

## Three sets

```python
for num in numbers:
    for perm in list(permutations(colors, 3)):
        first, second, third = perm
        temp = "{{Color: model."+first+", Number: model."+num+"}, {Color: model."+second+", Number: model."+num+"}, {Color: model."+third+", Number: model."+num+"}},"
        print(temp)
```

## Three sequence

```python
for c in colors:
    for i in range(len(numbers) - 2):
        first, second, third = numbers[i:i+3]
        temp="{{Color: model."+c+", Number: model."+first+"}, {Color: model."+c+", Number: model."+second+"}, {Color: model."+c+", Number: model."+third+"}},"
        print(temp)
```

## Four sequence

```python
for c in colors:
    for i in range(len(numbers) - 3):
        first, second, third, fourth = numbers[i:i+4]
        temp="{{Color: model."+c+", Number: model."+first+"}, {Color: model."+c+", Number: model."+second+"}, {Color: model."+c+", Number: model."+third+"}, {Color: model."+c+", Number: model."+fourth+"}},"
        print(temp)
```
